```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

print('Carregando Arquivo de teste')
arquivo = np.load('teste2.npy')
x = arquivo[0]
y = np.ravel(arquivo[1])





regr = MLPRegressor(hidden_layer_sizes=(2,4),
                    max_iter=700,
                    activation='relu', #{'identity', 'logistic', 'tanh', 'relu'},
                    solver='adam',
                    learning_rate = 'adaptive',
                    n_iter_no_change=50)
print('Treinando RNA')
regr = regr.fit(x,y)



print('Preditor')
y_est = regr.predict(x)


# Exemplo de valores de perda por época
#losses = [0.9, 0.8, 0.85, 0.7, 0.65, 0.68, 0.63, 0.6]

def best_loss(losses):
    min_loss = float('inf')  # Inicializa com infinito
    best_losses = []
    for loss in losses:
        if loss < min_loss:
            min_loss = loss
        best_losses.append(min_loss)
    return best_losses

# Gerando o melhor erro ao longo do tempo
#best_losses = best_loss(losses)

plt.figure(figsize=[14,7])

#losses1 = []
#plot curso original
for i in range(10):
    plt.subplot(1,3,1)
    plt.plot(x,y)
    #losses1 = MLPRegressor.best_loss_

#print(losses1)
#print("Media das perdas: ", np.min(losses1))


#plot aprendizagem
for i in range(10):
    plt.subplot(1,3,2)
    plt.plot(regr.loss_curve_)

#plot regressor
for i in range(10):
    plt.subplot(1,3,3)
    plt.plot(x,y,linewidth=1,color='yellow')
    plt.plot(x,y_est,linewidth=2)




plt.show()
```
